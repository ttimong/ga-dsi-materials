## Mean

```numpy.mean(a, axis=None, dtype=None)```

* a: array containing numbers whose mean is required
* axis: axis or axes along which the means are computed, 
    * default is to compute the mean of the flattened array
* dtype: type of data to be used in calculations

````python
print np.mean(B)
````

````python
print np.mean(B, axis=0)
````

````python
print np.mean(B, axis=1)
````

In the above code, axis=0 calculates the mean along every column and axis=1 calculates it along every row of the array.

````python
print np.mean(A) #this should give the same answer as np.mean(B). Why?
````

````python
print np.mean(A, axis=0) 
#this should give the same answer as np.mean(B, axis=1)
````

````python
print np.mean(A, axis=1) 
#this should give the same answer as np.mean(B, axis=0)
````

## Median
```numpy.median(a, axis=None, out=None)```

* a: array containing numbers whose median is required
* axis: axis or axes along which the median is computed
    * default is to compute the median of the flattened array
* out: alternative output array to place the result, must have the same shape and buffer length as the expected output.

````python
print A
````

````python
print B
````

````python
print np.median(B)
print np.median(A)
````

````python
print np.median(B, axis=0)
print np.median(A, axis=1)
````

````python
print np.median(B, axis=1)
print np.median(A, axis=0)
````

## Mode

```python
scipy.stats.mstats.mode(a, axis=0)
```

* a: array containing numbers whose mode is required
* axis: axis or axes along which the mode is computed
    * default is 0
    * if None, compute the mode of the flattened array

It returns (```mode: array of modal values, count: array of counts for each mode```).

````python
print A
````

````python
print B
````

````python
from scipy import stats
````

````python
print stats.mode(B)
print stats.mode(B,axis=0) # this is the default
print stats.mode(B,axis=None) # this is not the default
````

````python
print stats.mode(A)
print stats.mode(A,axis=0) # this is the default
print stats.mode(B,axis=None) # this is not the default
````

````python
print stats.mode(B,axis=0)
print stats.mode(A,axis=1)
````

````python
print stats.mode(B,axis=1)
print stats.mode(A,axis=0)
````

## Range


```python
numpy.ptp(a, axis=None, out=None)```

'ptp' stands for 'peak to peak'.
* a: array containing numbers whose range is required
* axis: axis or axes along which the range is computed, 
   * default is to compute the range of the flattened array. 

It returns a new array with the result.

````python
print A
````

````python
print B
````

````python
print np.ptp(B)
print np.ptp(A)
# the max value is 19
# the min value is 7
# range is max-min = 19-7 = 12
````

````python
print np.ptp(B,axis=0)
print np.ptp(A,axis=1)
````

````python
print np.ptp(B,axis=1)
print np.ptp(A,axis=0)
````

## Variance

Variance can be calculated in python using different libraries like numpy, pandas, and statistics.

```python
numpy.var(a, axis=None, dtype=None, ddof=0)
```

* a: array containing numbers whose variance is required
* axis: axis or axes along which the variances are computed, default is to compute the mean of the flattened array
* ddof : int, optional
    * ddof stands for delta degrees of freedom. It is the divisor used in the calculation, which is N - ddof, where N is the number of elements. 
    * The default value of ddof is 0.

````python
print A
````

````python
print B
````

````python
print np.var(B)
print np.var(A)
````

````python
print np.var(B,axis=0)
print np.var(A,axis=1)

print np.var(B,axis=1)
print np.var(A,axis=0)
````

## Standard Deviation

...the square root of variance.

It can be calculated in python using different libraries like numpy, pandas, and statistics.

```python
numpy.std(a, axis=None, dtype=None, ddof=0)
```
Parameters are the same as ```numpy.var()```.

* a: array containing numbers whose standard deviation is required
* axis: axis or axes along which the standard deviations are computed, default is to compute the mean of the flattened array
* ddof : int, optional
    * ddof stands for delta degrees of freedom. It is the divisor used in the calculation, which is N - ddof, where N is the number of elements. 
    * The default value of ddof is 0.

````python
print A
````

````python
print B
````

````python
print np.std(B)
print np.std(A)
````

````python
print np.std(B,axis=0)
print np.std(A,axis=1)

print np.std(B,axis=1)
print np.std(A,axis=0)
````

## Skewness

```python
scipy.stats.skew(a, axis=0)
```

For normally distributed data, the skewness should be about 0. A skewness value > 0 means that there is more weight in the left tail of the distribution.

````python
print B
````

````python
# from scipy import stats
````

````python
print stats.skew(B)
print stats.skew(B, axis=0)
print stats.skew(A, axis=1)

print stats.skew(B, axis=1)
print stats.skew(A, axis=0)
print stats.skew(A)

print stats.skew(B, axis=None)
print stats.skew(A, axis=None)
````

# Seaborn

````python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
sns.set(color_codes=True)

tips = sns.load_dataset("tips") #this dataset comes pre-loaded when you install seaborn
print type(tips)
print tips
````

````python
# Scatterplot arguments
sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, # No regression line
           hue='sex')   # Color by gender given in dataset
````

````python
# Scatterplot arguments
sns.lmplot(x='total_bill', y='tip', data=tips, markers=["x", "o"],palette="Set1",
           hue='sex')   # Color by gender given in dataset
````

````python
# Scatterplot arguments
sns.lmplot(x='total_bill', y='tip', data=tips, palette=dict(Male="b", Female="r"),
           hue='sex')   # Color by gender given in dataset
````

````python
g = sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
````

````python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
````

````python
g = sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, markers=["o", "x"])
````

````python
g = sns.lmplot(x="size", y="total_bill", hue="day", col="day",
...                data=tips, aspect=.4, x_jitter=.1)
````

````python
g = sns.lmplot(x="total_bill", y="tip", col="day", hue="day",
...                data=tips, col_wrap=2, size=3)
````

````python
g = sns.lmplot(x="total_bill", y="tip", row="sex", col="time",
...                data=tips, size=3)
````

````python
sns.boxplot(data=tips.drop(['size'], axis=1))
````
