### A. Explore _Mad Men_ Cast Data

---

#### 1. Load the _Mad Men_ cast data into a `pandas` DataFrame.

```python
cast_data_csv = './datasets/mad-men-cast-show-data.csv'
```

```python
cast_df = pd.read_csv(cast_data_csv)
```

#### 2. Print the head and tail of the data.

```python
cast_df.head()
```

#### 3. Print the columns of the data.

```python
list(cast_df.columns)
```

#### 4. Rename any columns with spaces or special characters to not contain any.

```python
import string
upper = string.ascii_uppercase
lower = string.ascii_lowercase

accp_char = upper+lower+' _'

new_column = [''.join([ch for ch in col if ch in accp_char]) for col in list(cast_df.columns)]
new_column = [col.lower() for col in new_column]
new_column = map(lambda x: x.replace(' ','_'), new_column)

cast_df.columns = new_column
```

#### 5. Subset the data where the status of the show is not "END" or "End."

```python
cast_df[(cast_df['status'] !="END") & (cast_df['status']!="End")]
```

#### 6. Print out the performers where the show start is greater than 2005 and the score is greater than 7.

```python
cast_df[(cast_df['show_start']) > 2005 & (cast_df['score'] >7)]['performer'].unique()
```

#### 7. Select the performer and show column for the 20th-25th LABELED rows.

```python
cast_df.loc[20:25,['performer', 'show']]
```

```python
cast_df.iloc[20:25, 0:2]
```

#### 8. Plot a histogram of score.

```python
cast_df['score'].hist()
```
